```python
import os
import cv2
from cv2.cv import *
```

```python
run_path = "./ucf/Archery/"
runList = os.listdir(run_path)
for vid in runList:
    fourcc = cv2.cv.CV_FOURCC(*'XVID')
    out = cv2.VideoWriter(vid,fourcc, 25.0, (112,112))
    capture = cv2.VideoCapture()
    success = capture.open(run_path+vid)
    if not success:
        print "Couldn't open video %s." % vid
    num_frames = int(capture.get(CV_CAP_PROP_FRAME_COUNT))
    for i in xrange(num_frames):
        frame_available, frame = capture.read()
        frame = cv2.resize(frame, (112,112)) 
        out.write(frame)
```
